```python
import pandas as pd
```

# CLEANING OF DATA

```python
df=pd.read_csv("all_merged_data.csv")
```

```python
df
```

```python
nan_df=df[df.isna().any(axis=1)]
nan_df
```

```python
df=df.dropna()
df
```

```python
df=df[df['Order Date'].str[0:2]!='Or']
df
```

```python
df["month"]=df["Order Date"].str[0:2]
df["month"]=df["month"].astype("int32")
df
```

```python
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'])
df['Price Each']=pd.to_numeric(df['Price Each'])
df
```

```python
df['sale']=df['Quantity Ordered']*df['Price Each']
df
```

```python
df.groupby('month').sum()
```

# visualising of Data

```python
import matplotlib.pyplot as plt
```

```python
result=df.groupby('month').sum()
```

```python
months=range(1,13)
plt.bar(months,result['sale'])
plt.xticks(months)
plt.ylabel("SALES IN MILLIONS")
plt.xlabel("MONTHS IN NUMBER")
plt.show()
```

# what time should we display advertisement to maximize likelihood of customers buying product

```python
df=df.drop(columns='city')
```

```python
df['Order Date']=pd.to_datetime(df["Order Date"])
```

```python
df["Hour"]=df["Order Date"].dt.hour
df["minute"]=df["Order Date"].dt.minute
df.head()
```
